```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import matplotlib.pyplot as mp
import seaborn as sns
```

```python
import matplotlib
```

```python
def maximum(x,y):
    max = np.argmax(y)
    return x[max]
```

```python
data1 = np.loadtxt("data.csv")
data1
```

```python
data = pd.DataFrame(data1)
data.head()
```

```python
x = data.iloc[:,1:]
y = data.iloc[:,:1]
#y.head()
```

# Q) 1) a)

```python
inertia = []
s =[]
n_clusters = range(2,21)
for i in n_clusters:
    model  = KMeans(n_clusters = i,init = 'random', n_jobs = -1)
    model.fit(x)
    inertia.append(model.inertia_)
    clusters = model.labels_
    s.append(silhouette_score(x,clusters))
```

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.scatter(n_clusters, inertia, c = 'r')
ax.plot(n_clusters, inertia)
plt.xlabel("No of Clusters")
plt.ylabel("Inertia")
plt.title("Figure 5.1")
```

# Q) 1) b)

# It seems like the optimal no of clusters is 12 from the Elbow approach

# Verifying using silhuotte score

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.scatter(n_clusters, s, c ='r')
ax.plot(n_clusters,s)
plt.xlabel("No of Clusters")
plt.ylabel("Silhuotte Score")
plt.title("Figure 5.1a")
```

```python
maximum(n_clusters, s)
```

# Yes. I am right. The optimal no of clusters = 12

# Q) 1) c)

```python
s2=[]
inertia2 = []
n_clusters = range(2,21)
for i in n_clusters:
    model  = KMeans(n_clusters = i,init = 'random', n_init = 1, max_iter = 1, n_jobs = -1)
    model.fit(x)
    inertia2.append(model.inertia_)
    clusters2 = model.labels_
    s2.append(silhouette_score(x,clusters2))
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.scatter(n_clusters, inertia2, c = 'r')
ax.plot(n_clusters, inertia2)
plt.xlabel("No of Clusters")
plt.ylabel("Inertia")
plt.title("Figure 5.2")
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.scatter(n_clusters, s2, c ='r')
ax.plot(n_clusters,s2)
plt.xlabel("No of Clusters")
plt.ylabel("Silhuotte Score")
plt.title("Figure 5.1a")
```

```python
maximum(n_clusters, s2)
```

# FOr this, the optimal no of clusters is 15

```python
iter = []
n_clusters = range(2,21)
for i in n_clusters:
    model  = KMeans(n_clusters = i,init = 'random',  n_jobs = -1)
    model.fit(x)
    iter.append(model.n_iter_)
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.scatter(n_clusters, iter, c = 'r')
ax.plot(n_clusters, iter)
plt.xlabel("No of Clusters")
plt.ylabel("Iterations")
plt.title("Figure 5.3")
```

# Graduate Student Question

```python
# Load data. csv file should be in the same folder as the notebook for this to work, otherwise
# give data path.
data_g = np.loadtxt("data.csv")
```

```python
#shuffle the data and select training and test data
np.random.seed(100)
np.random.shuffle(data_g)


features = []
digits = []


for row in data_g:
    #import the data and select only the 1's and 5's
    if(row[0]==1 or row[0]==5):
        features.append(row[1:])
        digits.append(str(row[0]))

        
#Select the proportion of data to use for training. 
#Notice that we have set aside 80% of the data for testing
numTrain = int(len(features)*.2)

trainFeatures = features[:numTrain]
testFeatures = features[numTrain:]
trainDigits = digits[:numTrain]
testDigits = digits[numTrain:]

```

```python
model_g  = KMeans(n_clusters = 10,init = 'random',  n_jobs = -1)
model_g.fit(x)
```

```python
labels_c = model_g.labels_
```

```python
#Convert the 256D data (trainFeatures) to 2D data
#We need X and Y for plotting and simpleTrain for building the model.
#They contain the same points in a different arrangement

X = []
Y = []
simpleTrain = []

#Colors will be passed to the graphing library to color the points.
#1's are blue: "b" and 5's are red: "r"
colors = []
for index in range(len(trainFeatures)):
    #produce the 2D dataset for graphing/training and scale the data so it is in the [-1,1] square
    xNew = 2*np.average(trainFeatures[index])+.75 
    yNew = 3*np.var(trainFeatures[index])-1.5
    X.append(xNew)
    Y.append(yNew)
    simpleTrain.append([xNew,yNew])
    #trainDigits will still be the value we try to classify. Here it is the string "1.0" or "5.0"
    if(trainDigits[index]=="1.0"):
        colors.append("b")
    else:
        colors.append("r")

#plot the data points
### https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html
mp.scatter(X,Y,s=3,c=colors)

#specify the axes
mp.xlim(-1,1)
mp.xlabel("Average Intensity")
mp.ylim(-1,1)
mp.ylabel("Intensity Variance")
mp.title("Figure 1.1")
#display the current graph
mp.show()
for i in range(len(simpleTrain)):
    simpleTrain[i].append(labels_c[i])
    
```

```python
data_c = pd.DataFrame(data = simpleTrain, columns = ['X', 'Y', 'Label'])
data_c.head()
```

```python
color_labels = data_c['Label'].unique()

# List of RGB triplets
rgb_values = sns.color_palette("Set2", 10)

# Map label to RGB
color_map = dict(zip(color_labels, rgb_values))
plt.figure(figsize=(10, 5))
# Finally use the mapped values
plt.scatter(data_c['X'], data_c['Y'], c=data_c['Label'].map(color_map))
```

```python
model_e = KMeans(n_clusters = 10, init = 'random')
model_e.fit(x)
clusters_e = model_e.labels_
metrics.adjusted_rand_score(y[0],clusters_e)
```

```python
acc =[]
n_clusters = range(2,33)
for i in n_clusters:
    model3  = KMeans(n_clusters = i,init = 'random', n_jobs = -1)
    model3.fit(x)
    clusters3 = model3.labels_
    acc.append(metrics.adjusted_rand_score(y[0],clusters3))
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.scatter(n_clusters, acc, c ='r')
ax.plot(n_clusters,acc)
plt.xlabel("No of Clusters")
plt.ylabel("Accuracy")
plt.title("Figure")
```

```python
maximum(n_clusters, acc)
```
